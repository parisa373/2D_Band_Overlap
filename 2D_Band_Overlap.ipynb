```python
import matplotlib.pyplot as plt

import numpy as np

k_x = 0
a = 1
h = 1
m = 1
coef = (h**2)/(2*m)

k_y = np.linspace(-np.pi/a, np.pi/a, 1000)

E = coef*(k_y**2)
E2 = coef*((k_y-2*np.pi/a)**2)
E3 = coef*((k_y+2*np.pi/a)**2)


plt.plot(k_y, E, color = 'blue')
plt.plot(k_y, E2, color = 'red')
plt.plot(k_y, E3, color = 'red')


```

```python

a = 1
h = 1
m = 1
coef = (h**2)/(2*m)
k_x1 = np.pi/a

k_y = np.linspace(-np.pi/a, np.pi/a, 1000)

E4 = coef*(k_y**2 + k_x1**2)
E5 = coef*((k_y-2*np.pi/a)**2 + k_x1**2)
E6 = coef*((k_y+2*np.pi/a)**2 + k_x1**2)

plt.plot(k_y, E4, color = 'green')
plt.plot(k_y, E5, color = 'violet')
plt.plot(k_y, E6, color = 'violet')


```

```python
a = 1
h = 1
m = 1
coef = (h**2)/(2*m)
k_x2 = np.linspace(-np.pi/a, np.pi/a, 1000)
k_y2 = np.linspace(-np.pi/a, np.pi/a, 1000)

k1 = np.sqrt(k_x2**2 + k_y2**2)
E7 = coef*(k1**2)

k2 = np.sqrt((k_x2-2*np.pi/a)**2 + (k_y2-2*np.pi/a)**2)
E8 = coef*(k2**2)

k3 = np.sqrt((k_x2+2*np.pi/a)**2 + (k_y2+2*np.pi/a)**2)
E9 = coef*(k3**2)

plt.plot(k1, E7, color='orange')
plt.plot(k2, E8, color='black')
plt.plot(k3, E9, color='yellow')
```
